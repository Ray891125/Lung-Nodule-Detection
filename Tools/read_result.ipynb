```python

import os
import re
import numpy as np

dist= r"SANet_npatch_focal01\SANet\cross_val_test_Dicom_crop\201\res"
Base_save_dir = r"C:\Users\BB\Desktop\training_data\My_SAnet\Save"
save_dir = os.path.join(Base_save_dir,dist)
folders_list = [folder for folder in os.listdir(save_dir)]
folders = []

for folder in folders_list:
    try:
        folder_int = int(folder)
        if type(folder_int)==int:
            folders.append(folder_int)
    except ValueError:
        print(f"Skipping folder '{folder}' due to conversion error.")
folders=sorted(folders)

for folder in folders:
    Analysis_dir = os.path.join(save_dir, str(folder), 'FROC', 'iou', 'rpn', 'Analysis_0.1.txt')  
    with open(Analysis_dir, 'r') as file:
        file_content = file.read()
        sensitivity_pattern = re.compile(r'Sensitivity:\s+(\d+\.\d+)')
        sensitivity_match = sensitivity_pattern.search(file_content)

        if sensitivity_match:
            print(folder)
            sensitivity_value = sensitivity_match.group(1)
            print("Sensitivity value:", sensitivity_value)
        else:
            print("Sensitivity value not found.")

        # 使用正则表达式匹配FP/Scan, Sensitivity, Precision部分
        fp_scan_pattern = re.compile(r'FP/Scan,\s+Sensitivity,\s+Precision\n([\d\.,\s]+)')
        fp_scan_match = fp_scan_pattern.search(file_content)

        if fp_scan_match:
            fp_scan_values = fp_scan_match.group(1)
            print("FP/Scan, Sensitivity, Precision values:")
            print(fp_scan_values)
        else:
            print("FP/Scan, Sensitivity, Precision values not found.")

        # 使用正则表达式匹配Froc_mean的数值
        froc_mean_pattern = re.compile(r'Froc_mean = (\d+\.\d+)')
        froc_mean_match = froc_mean_pattern.search(file_content)

        if froc_mean_match:
            froc_mean_value = froc_mean_match.group(1)
            print("Froc_mean value:", froc_mean_value)
            print("-----------------------------------")
        else:
            print("Froc_mean value not found.")
        
```
